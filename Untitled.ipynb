```python
df = pd.read_csv("RAVDESS.csv")

df = df[df.label != 2]
df_clean = df.dropna()

X = df_clean.drop(columns=["label"]).values.astype(float)
Y = df_clean["label"].values

le = LabelEncoder()
y_encoded = le.fit_transform(Y)

scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, stratify=y_encoded, test_size=0.2, random_state=42)
print(X_train.shape)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_ds = TensorDataset(X_train_tensor, y_train_tensor)
train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
print(train_dl)
```

```python
class EmotionNet(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(EmotionNet, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        return self.model(x)

model = EmotionNet(input_dim=X.shape[1], num_classes=len(le.classes_))
```

```python
criterion = nn.CrossEntropyLoss()
opti = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(2):
    model.train()
    total_loss = 0
    for xb, yb in train_dl:
        opti.zero_grad()
        pred = model(xb)
        loss = criterion(pred, yb)
        loss.backward()
        opti.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}: Loss = {total_loss/len(train_dl):.4f}")
```

```python
model.eval()

with torch.no_grad():
    preds = model(X_test_tensor)
    y_pred = torch.argmax(preds, axis=1)

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
cm_mat = confusion_matrix(y_test, y_pred.numpy())
cm = ConfusionMatrixDisplay(confusion_matrix=cm_mat)

cm.plot()
plt.show()

target_names = [str(cls) for cls in le.classes_]
print(classification_report(y_test, y_pred.numpy(), target_names=target_names))
```

```python
def mfcc_extract(audio_file, limit=100):
    print(audio_file)
    y, sr = librosa.load(audio_file)
    y_trim, _ = librosa.effects.trim(y, top_db=20)
    
    if librosa.get_duration(y=y_trim, sr=sr) < 0.5: 
        return None

    mfcc = librosa.feature.mfcc(y=y_trim, sr=sr, n_mfcc=15)

    if mfcc.shape[1] < limit:
        pad = limit - mfcc.shape[1]
        padded_mfcc = np.pad(mfcc, pad_width=((0,0),(0,pad)))
    else:
        padded_mfcc = mfcc[:, :limit]     

    return padded_mfcc

audio_files = glob("TESS/TESS Toronto emotional speech set data/OAF_Pleasant_surprise/*.wav")

feature_list = []
for file in audio_files:
    vec = mfcc_extract(file)
    # label = os.path.basename(file).split("-")[2]
    label = 8
    
    filename = os.path.splitext(os.path.basename(file))[0]
    save_path = f"numpy_arrays/{filename}.npz"
    
    np.savez(save_path, mfcc=vec, label=label)
    feature_list.append([save_path, label])

df = pd.DataFrame(feature_list, columns=["array_location", "labels"])
df.to_csv("emotion.csv", mode='a', index=False, header=False)
print("success")
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)

X = (X - X.mean(axis=(0, 1), keepdims=True)) / (X.std(axis=(0, 1), keepdims=True) + 1e-8)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, stratify=y_encoded, test_size=0.2, random_state=42)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).permute(0, 2, 1)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32).permute(0, 2, 1)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_ds = TensorDataset(X_train_tensor, y_train_tensor)
train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
print(train_dl)
```

```python
class EmoCnn(nn.Module):
    def __init__(self, num_classes):
        super(EmoCnn, self).__init__()
        self.conv1 = nn.Conv1d(100, 64, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm1d(64)
        self.pool1 = nn.MaxPool1d(2)

        self.conv2 = nn.Conv1d(64, 128, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm1d(128)
        self.pool2 = nn.MaxPool1d(2)

        self.lstm = nn.LSTM(128, 64, batch_first=True, bidirectional=True)

        self.l1 = nn.Linear(2*64, 128)
        self.dropout = nn.Dropout(0.3)
        self.l2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))

        x = x.permute(0, 2, 1)

        lstm_out, _ = self.lstm(x)
        last_out = lstm_out[:, -1, :]

        x = F.relu(self.l1(last_out))
        x = self.dropout(x)
        x = self.l2(x)

        return x

model = EmoCnn(len(le.classes_))
model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
opti = torch.optim.Adam(model.parameters(), lr=0.00001)

for epoch in range(5):
    model.train()
    total_loss = 0
    for xb, yb in train_dl:
        xb, yb = xb.to(device), yb.to(device)
        opti.zero_grad()
        pred = model(xb)
        loss = criterion(pred, yb)
        loss.backward()
        opti.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}: Loss = {total_loss/len(train_dl):.4f}")
```

```python
model.eval()
X_test_tensor = X_test_tensor.to(device)
y_test_tensor = y_test_tensor.to(device)

with torch.no_grad():
    preds = model(X_test_tensor)
    y_pred = torch.argmax(preds, axis=1)

# targets = ["neutral", "happy", "sad", "angry", "fearful", "disgust", "surprised"]

from sklearn.metrics import classification_report
print(classification_report(y_test_tensor.cpu().numpy(), y_pred.cpu().numpy()))
```

```python
model = EmoCnn(len(le.classes_))
model.load_state_dict(torch.load("cnn_model.pth", weights_only=True))
model.to(device)
model.eval()

def emo_model(audio_file):
    y, sr = librosa.load(audio_file)
    y_trim, _ = librosa.effects.trim(y, top_db=20)
    
    if librosa.get_duration(y=y_trim, sr=sr) < 0.5: 
        return None

    mfcc = librosa.feature.mfcc(y=y_trim, sr=sr, n_mfcc=15)

    mfcc_mean = np.mean(mfcc, axis=1, keepdims=True)
    mfcc_std = np.std(mfcc, axis=1, keepdims=True)
    mfcc_std[mfcc_std == 0] = 1e-8  
    mfcc = (mfcc - mfcc_mean) / mfcc_std

    if mfcc.shape[1] < 100:
        pad = 100 - mfcc.shape[1]
        padded_mfcc = np.pad(mfcc, pad_width=((0,0),(0,pad)))
    else:
        padded_mfcc = mfcc[:, :100]   
    
    tensor_input = torch.tensor(padded_mfcc, dtype=torch.float32).unsqueeze(0).permute(0, 2, 1).to(device)

    with torch.no_grad():
        out = model(tensor_input)
        pred = torch.argmax(out, dim=1).item()
    
    emo_dict = {
        0: "neutral",
        1: "happy",
        2: "sad",
        3: "angry",
        4: "fearful",
        5: "disgust",
        6: "surprised"
    }

    return emo_dict[pred]


import gradio as gr
demo = gr.Interface(emo_model, inputs=gr.Audio(type="filepath"), outputs=gr.Textbox(label="Predicted Emotion"))
demo.launch()
```
